```python
import numpy as np
import matplotlib.pyplot as plt


class LinearRegression:

    def __init__(self):
        self.m = None
        self.b = None

    def fit(self, x_train, y_train):

        num = 0
        den = 0

        for i in range(x_train.shape[0]):

            num = num + (
                (x_train[i] - x_train.mean()) *
                (y_train[i] - y_train.mean())
            )

            den = den + (
                (x_train[i] - x_train.mean()) *
                (x_train[i] - x_train.mean())
            )

        self.m = num / den

        self.b = y_train.mean() - (self.m * x_train.mean())

        print("Slope (m) =", self.m)
        print("Intercept (b) =", self.b)

    def predict(self, x_test):

        return self.m * x_test + self.b

    def plot_graph(self, x_train, y_train):

        y_pred = self.m * x_train + self.b

        plt.scatter(x_train, y_train, label="Actual Data")

        plt.plot(x_train, y_pred, label="Regression Line")

        plt.xlabel("X")
        plt.ylabel("Y")

        plt.title("Simple Linear Regression")

        plt.legend()

        plt.show()


# Training Data
x_train = np.array([1, 2, 3, 4, 5])
y_train = np.array([2, 4, 6, 8, 10])

# Create Model
model = LinearRegression()

# Train Model
model.fit(x_train, y_train)

# Prediction
result = model.predict(6)

print("Prediction for X=6 :", result)

# Draw Graph
model.plot_graph(x_train, y_train)
```
